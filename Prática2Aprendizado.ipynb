```
import pandas as pd
import matplotlib.pyplot as plt

wine_data_path = 'wine.data'
wine_names_path = 'wine.names'

columns = [
    "Class", "Alcohol", "Malic_Acid", "Ash", "Alcalinity_of_Ash",
    "Magnesium", "Total_Phenols", "Flavanoids", "Nonflavanoid_Phenols",
    "Proanthocyanins", "Color_Intensity", "Hue", "OD280_OD315", "Proline"
]

df = pd.read_csv(wine_data_path, header=None, names=columns)

class_column = 'Class'

class_counts = df[class_column].value_counts()
class_counts.plot.pie(
    autopct='%1.1f%%',
    startangle=90,
    colors=['#FF9999', '#66B2FF', '#99FF99'],
    labels=[f'Class {int(c)}' for c in class_counts.index]
)

plt.title('Distribuição das Classes de Vinho')
plt.ylabel('')
plt.show()


print(f"Distribuição das classes:\n{class_counts}")
```

```
import pandas as pd

wine_data = pd.read_csv("wine.data", header=None)

wine_data.columns = ['Class', 'Alcohol', 'Malic_acid', 'Ash', 'Alcalinity_of_ash',
                     'Magnesium', 'Total_phenols', 'Flavanoids', 'Nonflavanoid_phenols',
                     'Proanthocyanins', 'Color_intensity', 'Hue',
                     'OD280/OD315_of_diluted_wines', 'Proline']

missing_data = wine_data.isna().sum()

print("Dados faltantes por coluna:")
print(missing_data)
```

```
import seaborn as sns
import matplotlib.pyplot as plt

numerical_features = wine_data.select_dtypes(include=['float64', 'int64'])

sns.pairplot(numerical_features)
plt.suptitle('Análise Multivariada das Features Numéricas', y=1.02)
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = wine_data.select_dtypes(include=['float64', 'int64']).corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlação de Pearson entre as Features Numéricas', fontsize=16)
plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

numerical_columns = wine_data.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x='Class', y=col, data=wine_data)
    plt.title(f'Boxplot de {col} por Classe')
    plt.show()

    Q1 = wine_data[col].quantile(0.25)
    Q2 = wine_data[col].quantile(0.50)
    Q3 = wine_data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    print(f"\nPara a feature {col}:")
    print(f"Q1: {Q1}")
    print(f"Q2 (Mediana): {Q2}")
    print(f"Q3: {Q3}")
    print(f"IQR: {IQR}")
    print(f"Limite Inferior: {lower_limit}")
    print(f"Limite Superior: {upper_limit}")
```

```
from sklearn.impute import SimpleImputer
import seaborn as sns
import matplotlib.pyplot as plt

wine_data_imputed = wine_data.copy()

imputer = SimpleImputer(strategy='median')
wine_data_imputed[numerical_columns] = imputer.fit_transform(wine_data[numerical_columns])

for col in numerical_columns:
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    sns.boxplot(x='Class', y=col, data=wine_data)
    plt.title(f'Boxplot de {col} - Dados Brutos')

    plt.subplot(1, 2, 2)
    sns.boxplot(x='Class', y=col, data=wine_data_imputed)
    plt.title(f'Boxplot de {col} - Dados Imputados')

    plt.tight_layout()
    plt.show()
```
